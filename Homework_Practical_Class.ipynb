```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Input Data
# Years of Experience
X = np.array([[1], [2], [3], [4], [5]])

# Salaries
y = np.array([30000, 35000, 40000, 45000, 50000])

# Step 2: Create the model
model = LinearRegression()

# TODO: Train the model
model.fit(X, y)  #(fill in the blank)

# Step 3: Predict a salary
# TODO: Predict salary for 6 years experience
years = 6
predicted_salary = model.predict([[years]])
print("Predicted salary for 6 years of experience:", predicted_salary[0])
```

```python
# TODO: Now plot this graph with the predicted value displayed as a green point
# Plotting
import matplotlib.pyplot as plt

plt.scatter(X, y, color='black', label='Actual Salaries')  # Plot actual data points
plt.plot(X, model.predict(X), color='orangered', label='Regression Line')  # Plot regression line
plt.scatter(years,predicted_salary, color='green', marker='o', s=100, label='Predicted Salary for 6 Years')  # Plot predicted point

# Adding labels and title
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()


# Show the plot
plt.show()
```
